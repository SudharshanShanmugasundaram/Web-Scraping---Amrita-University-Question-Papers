```python
import requests
from bs4 import BeautifulSoup
import webbrowser
from prettytable import PrettyTable
import os

def func4(url,URL):
    n=PrettyTable(['SNo.','Available Question Papers'])
    page=requests.get(url)
    data=BeautifulSoup(page.content,'html.parser')
    div=data.findAll('div',class_='file-list')
    div_file=div[0].findAll('div',class_='file-wrapper clearfix')
    for i in range(len(div_file)):
        div_style=div_file[i].findAll('div',class_='file-metadata')
        div_span=div_style[0].findAll('div')
        span=div_span[0].findAll('span')
        name=span[1].text
        n.add_row([i+1,name])
    print(n)
    ch=int(input("Enter corresponding serial number to download/view the file "))
    url1=URL+div_file[ch-1].a['href']
    webbrowser.open(url1)
    
def func3(url,URL):
    m=PrettyTable(['SNo.',' More Collection'])
    page=requests.get(url)
    data=BeautifulSoup(page.content,'html.parser')
    div=data.findAll('div',class_='ds-static-div secondary recent-submission')
    ul=div[0].findAll('ul')
    li=ul[0].findAll('li')
    for i in range(len(li)):
        name=li[i].a.text
        m.add_row([i+1,name])
    m.align[' More Collection']='l'
    print(m)
    ch=int(input("Enter appropriate number to view question papers "))
    url1=URL+li[ch-1].a['href']
    func4(url1,URL) 
    
def func2(url,URL):
    c=PrettyTable(['SNo.','Collection'])
    page=requests.get(url)
    data=BeautifulSoup(page.content,'html.parser')
    div=data.findAll('div',class_='ds-static-div secondary')
    ul=div[0].findAll('ul')
    li=ul[1].findAll('li')
    for i in range(len(li)):
        name=li[i].a.text
        c.add_row([i+1,name])
    c.align['Collection']='l'
    print(c)
    ch=int(input("Enter appropriate number to view more collections "))
    url1=URL+li[ch-1].a['href']
    func3(url1,URL)

def func1(url,URL):
    s=PrettyTable(['SNo.','Semester'])
    page=requests.get(url)
    data=BeautifulSoup(page.content,'html.parser')
    div=data.findAll('div',class_='ds-static-div secondary')
    ul=div[0].findAll('ul')
    li=ul[0].findAll('li')
    for i in range(len(li)):
        name=li[i].a.text
        s.add_row([i+1,name])
    s.align['Semester']='l'
    print(s)
    ch=int(input("Enter appropriate number to view the collections "))
    url1=URL+li[ch-1].a['href']
    func2(url1,URL)
    
URL='http://dspace.amritanet.edu:8080'    
url="{}/xmlui/handle/123456789/16".format(URL)
url1='{}'.format(URL)
d=PrettyTable(['SNo.','Department'])
page=requests.get(url)
data=BeautifulSoup(page.content,'html.parser')
div=data.findAll('div',class_='ds-static-div secondary')
ul=div[0].findAll('ul')
li=ul[0].findAll('li')

for i in range(len(li)):
    name=li[i].a.text
    d.add_row([i+1,name])
d.align['Department']='l'
print(d)

ch=int(input("Enter appropriate number to view semesters "))
url1+=li[ch-1].a['href']
func1(url1,URL)
```
